```python
"""
Author: Mario Simental
ID: 3049387
Assignment: Assignment 2a
Class: EECS690
"""

import pandas as pd # Import pandas
import altair as alt # Import altair

# Load datasets
hp_report = pd.read_csv("./data/happiness_report.csv") # Load csv file
hp_report_semi_colon = pd.read_csv("./data/happiness_report_semicolon.csv", sep=";") # Load semi-colon seperated file
hp_report_meta_data = pd.read_csv("./data/happiness_report_metadata.csv", skiprows=2) # Load meta-data file
hp_report_xlsx = pd.read_excel("./data/happiness_report.xlsx") # Load spreadsheet file
happy_header = pd.read_csv("./data/happiness_report_no_header.csv", names=["country", "happiness_score", "GDP_per_capita", "life_expectancy", "freedom"]) # Load no header file and add headers

# Print datasets
print(f"Comma\n{hp_report}") # Print csv file
print(f"Semi Colon\n{hp_report_semi_colon}") # Print semi-colon seperated file
print(f"Meta Data\n{hp_report_meta_data}") # Print meta-data file
print(f"Spreadsheet\n{hp_report_xlsx}") # Print spreadsheet file
print(f"Happy header\n{happy_header}") # Print no header file

# Scatter plot: life_expectancy vs. GDP_per_capita
scatter_plot = alt.Chart(happy_header).mark_point().encode(
    x=alt.X('GDP_per_capita', title='How Much GDP Contributes To Happiness Score'),
    y=alt.Y('life_expectancy', title='How Much Life Expectancy Contribtes To Happiness Score'),
).configure_axis(
    labelFontSize=12,
    titleFontSize=12
)

scatter_plot # Display scatter plot

# Bar chart: Prepare a meaningful bar chart using metadata
bar_chart = alt.Chart(hp_report_meta_data).mark_bar().encode(
    x=alt.X('happiness_score:Q', bin=alt.Bin(maxbins=11), title='happiness_score(binned)'), 
    y=alt.Y('count():Q', title='Count of Records'),
  
).configure_axis(
    labelFontSize=12,
    titleFontSize=12
)

bar_chart # Display bar chart

```
