```python
"""
Author: Mario Simental
ID: 3049387
Assignment: Assignment 3
Class: EECS690
"""

import pandas as pd
import altair as alt

# Task 1.1

# Enable VegaFusion to raise limit of dataset (THIS WAS NECCESSARY ON MY PC BUT NOT LAPTOP)
#alt.data_transformers.enable("vegafusion")

# Load the data
avocado_prices = pd.read_csv("data/avocado_prices.csv")

# Display the first few rows of the data frame
avocado_prices
```

```python
# Task 1.2

# Convert the 'Date' column to datetime
avocado_prices["Date"] = pd.to_datetime(avocado_prices["Date"])

# Define the start and end dates for the X-axis
start_date = "2015-01-01" # This is the exact date as shown in the example
end_date = "2015-11-02" # This is the exact date as shown in the example + 1 extra day

# Filter the data to include only rows within the specified date range
filtered_data = avocado_prices[
    (avocado_prices["Date"] >= start_date) & (avocado_prices["Date"] <= end_date)
]

# Create the chart with filtered data
chart = alt.Chart(filtered_data).mark_point().encode(
    x=alt.X("Date:T", title="Date", # Set the X-axis
            scale=alt.Scale(domain=[start_date, end_date]),  # Set the X-axis range
            axis=alt.Axis(format="%Y-%m-%d", labelAngle=-90)),  # Format and rotate labels
    y=alt.Y("average_price:Q", title="Average Price", # Set the Y-axis
            scale=alt.Scale(domain=[0, 3.5]))  # Average price on the Y-axis
).properties(
    width=1000  # Stretch the chart horizontally
)

# Display the chart
chart
```

```python
# Task 1.3

# Group by week and compute the average price for each week
avocado_prices_weekly = avocado_prices.groupby("wk", as_index=False).agg({
    "average_price": "mean",
    "small_hass_volume": "mean",
    "large_hass_volume": "mean",
    "extra_l_hass_volume": "mean",
    "yr": "mean"
})

# Display the first 5 results
avocado_prices_weekly.head(5)
```

```python
# Task 1.4

# Create the chart with filtered data
chart_weekly = alt.Chart(avocado_prices_weekly).mark_point().encode(
    x=alt.X("wk:Q", title="Week"), # Set the X-axis
    y=alt.Y("average_price:Q", title="Average Price (US Dollars)", scale=alt.Scale(domain=[1.2, 1.65])), # Set the Y-axis
)

# Display the chart
chart_weekly
```

```python
# Task 1.5

# Calculate the total volume
avocado_prices["total_volume"] = (
    avocado_prices["small_hass_volume"] + 
    avocado_prices["large_hass_volume"] + 
    avocado_prices["extra_l_hass_volume"]
)

# Display the first few results
avocado_prices
```

```python
# Task 1.6

# Group by week and compute the average price for each week
avocado_aggregate_2 = avocado_prices.groupby("wk", as_index=False).agg({
    "average_price": "mean",
    "small_hass_volume": "mean",
    "large_hass_volume": "mean",
    "extra_l_hass_volume": "mean",
    "yr": "mean",
})

# Calculate the total volume
avocado_aggregate_2["total_volume"] = (
    avocado_aggregate_2["small_hass_volume"] + 
    avocado_aggregate_2["large_hass_volume"] + 
    avocado_aggregate_2["extra_l_hass_volume"]
)

# Display the first 5 results
avocado_aggregate_2.head(5)
```

```python
# Task 1.7

# Create the chart with filtered data
avocado_aggregate_plot_2 = alt.Chart(avocado_aggregate_2).mark_point().encode(
    x=alt.X("wk:Q", title="Week"), # Set the X-axis
    y=alt.Y("total_volume:Q", title="Total Volume", scale=alt.Scale(domain=[250_000, 600_000])), # Set the Y-axis
)

# Display the chart
avocado_aggregate_plot_2
```

```python
# Task 2.1

# Load the data
departure_bay_temperature = pd.read_csv("data/departure_bay_temperature.csv", skiprows=2)

# Display the first few rows of the data frame
departure_bay_temperature
```

```python
# Task 2.2

# Tidy up the data: wide to long
departure_bay_temperature_tidy = departure_bay_temperature.melt(id_vars="Year", var_name="Month", value_name="Temperature")

# Display the first few rows of the data frame
departure_bay_temperature_tidy
```

```python
# Task 2.3

# Filter the data where the month is November
departure_bay_temperature_november = departure_bay_temperature_tidy[departure_bay_temperature_tidy["Month"] == "Nov"]

# Create the chart for the new data
departure_bay_temperature_chart = alt.Chart(departure_bay_temperature_november).mark_point().encode(
    x=alt.X("Year:Q", title="Year", scale=alt.Scale(domain=[1905, 2015])), # Create the X-axis
    y=alt.Y("Temperature:Q", title="Temperature (Celsius)", scale=alt.Scale(domain=[4, 9])) # Create the Y-axis
)

# Display the chart for november
departure_bay_temperature_chart
```

```python
# Task 2.4

# Create the faceted chart for all months
all_temp_plot = alt.Chart(departure_bay_temperature_tidy).mark_point().encode(
    x=alt.X("Year").title("Year").scale(zero=False),
    y=alt.Y("Temperature").title("Temperature (Celsius)").scale(zero=False)
).facet("Month", columns=4)

# Display the faceted chart
all_temp_plot
```
