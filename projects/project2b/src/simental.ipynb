```python
"""
Author: Mario Simental
ID: 3049387
Assignment: Assignment 2b
Class: EECS690
"""

import ibis # Import ibis
import altair as alt # Import altair

# Connect to the SQLite database
conn = ibis.sqlite.connect("data/flights_filtered.db")

# List tables in the database and select bos_flights table
print(conn.list_tables())
flights_table_name = "bos_flights" 

# Create a reference to the table
flight_data = conn.table(flights_table_name)
print(flight_data)  # Print the table schema

# Inspect the first few rows of the table
print(flight_data.head()) # Print the first few rows 
print(flight_data.head().execute()) # Print the formatted result

# Filter rows where the origin airport is "BOS" and select ARRIVAL_DELAY and DEPARTURE_DELAY columns
fd_bos_origin = flight_data.filter(flight_data["ORIGIN_AIRPORT"] == "BOS").execute()
delay_data = fd_bos_origin[["ARRIVAL_DELAY", "DEPARTURE_DELAY"]]

# Inspect the filtered data
print(delay_data) # Print the filtered data
print(delay_data.shape) # Print the shape of the filtered data

# Convert delay times from minutes to hours
delay_data = delay_data.assign(
    ARRIVAL_DELAY_hr=delay_data["ARRIVAL_DELAY"] / 60,
    DEPARTURE_DELAY_hr=delay_data["DEPARTURE_DELAY"] / 60
)

# Plot arrival delay histogram
arrival_delay_plot = alt.Chart(delay_data).transform_joinaggregate(
    total="count(ARRIVAL_DELAY_hr)"
).transform_calculate(
    pct="1 / datum.total"
).mark_bar().encode(
    alt.X("ARRIVAL_DELAY_hr:Q")
        .bin(step=0.25, extent=[-2, 5])
        .title("Delay (hours)"),
    alt.Y("sum(pct):Q")
        .axis(format="%")
        .title("% of Flights")
)
arrival_delay_plot.show() # Display the plot

# Plot departure delay histogram
departure_delay_plot = alt.Chart(delay_data).transform_joinaggregate(
    total="count(DEPARTURE_DELAY_hr)"
).transform_calculate(
    pct="1 / datum.total"
).mark_bar().encode(
    alt.X("DEPARTURE_DELAY_hr:Q")
        .bin(step=0.25, extent=[-2, 5])
        .title("Delay (hours)"),
    alt.Y("sum(pct):Q")
        .axis(format="%")
        .title("% of Flights")
)
departure_delay_plot.show() # Display the plot

# Save the filtered data to a CSV file
delay_data.to_csv("data/delay_data.csv", index=False)
```
